```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
from datetime import datetime
from sklearn import preprocessing
```

```python
def date_transform(data):
    df = data.copy()
    
    df['Hour'] = df.index.hour
    df['Dayofweek'] = df.index.dayofweek
    df['Dayofmonth'] = df.index.day
    df['Dayofyear'] = df.index.dayofyear
    df['Month'] = df.index.month
    df['Quarter'] = df.index.quarter
    df['Year'] = df.index.year
    
    X = df
    
    return X
```

## Next winter time

```python
min_date_train = '2022-10-01'
max_date_train = '2023-04-30'
format_date = "%Y-%m-%d"
min_date_obj = pd.to_datetime(min_date_train)
min_date_obj
```

```python
model_folder = "../model/"
```

```python
result = pd.DataFrame(columns=['address_uuid', 'date', 'volume'])
```

```python
house_characteristics = pd.read_csv("../data/характеристики_домов.csv", delimiter=";")
```

```python
all_address_uuid = house_characteristics["address_uuid"].unique()
```

```python
for house_address in all_address_uuid:
    try: 
        model = joblib.load(model_folder + 'xgb_model' + house_address + '.sav')

        # prepare dataset
        test_df = pd.DataFrame(pd.date_range(start=min_date_train, end=max_date_train))
        test_df.rename(columns={0: "date"}, inplace=True)
        test_df.set_index("date", inplace=True)
        test_df = date_transform(test_df)

        predicted = model.predict(test_df)
        
        for i in range(len(predicted)):
            row = [house_address, min_date_obj + pd.DateOffset(i), predicted[i]]
            result.loc[len(result)] = row
    except FileNotFoundError as e:
        print(e)
```

```python
output_path = "../result/"
result.to_csv(output_path + 'volume.csv', index=False, encoding='utf-8', sep=";")
```
