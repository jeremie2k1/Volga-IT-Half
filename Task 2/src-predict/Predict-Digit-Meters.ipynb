```python
import os
from os import listdir
import shutil
import numpy as np
from matplotlib import pyplot as plt
import cv2
import pandas as pd
import joblib
```

```python
from roboflow import Roboflow
```

```python
source_data = '../../../Tasks/2/Data/'
cropped_data = '../../../Tasks/2/Reading-Meters-Data/'
```

```python
# Load model
model_detect_digit = joblib.load("../model/model_digital_number.sav")
```

```python
# dataframe to write csv file
result = pd.DataFrame(columns=['image_id', 'data1', 'data2'])
```

```python
volga_images_data = np.array([])
for images in os.listdir(source_data):
    if (images.endswith(".jpg")):
        volga_images_data = np.append(volga_images_data, images)
volga_images_data.shape
```

## Predict by parallel

```python
import multiprocessing
```

```python
# @background
async def detect_digit(image):
    cropped_path = cropped_data + image
    cropped_image_zoomin = cv2.imread(cropped_path)
    res = []
    try:
        digit_rgnz = model_detect_digit.predict(cropped_image_zoomin, confidence=50, overlap=30).json()['predictions']
        sorted_digit_rgnz = sorted(digit_rgnz, key=lambda d: (d['x'], d['y']))
        # linear_angle = cal_linear_angle(sorted_digit_rgnz)
        # print(image)
        for digit in sorted_digit_rgnz:
            if digit['class'].isdigit():
                res.append(digit['class'])
            # print(digit['class'], end=' ')
        print()
        
    except:
        unrecognize_number.append(image)
        return res
    return res
```

```python
unrecognize_number = []
```

```python
async def handed_detected(detected):
    if len(detected) > 5:
        return ["".join(map(str, detected[:5])), "".join(map(str, detected[5:]))]
    else:
        return ["".join(map(str, detected)), ""]
```

```python
for image in volga_images_data:
    res = [image]
    detected = await detect_digit(image)
    res.extend(await handed_detected(detected))
    print(res)
    print('-' * 20)
    result.loc[len(result)] = res
```

## Write to csv file

```python
output_path = "../result/"
result.to_csv(output_path + 'values.csv', index=False, encoding='utf-8', sep=";")
```
